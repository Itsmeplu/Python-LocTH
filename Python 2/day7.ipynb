## openings

r = read
w = write
a = append
x = create
w+ = write and read
a+ = apeend and read
x+ = create and read

```python
f = open('hello.txt', 'a+')
```

```python
f.seek(0)
print(f.read())
```

Seekers

```python
f = open('hello.txt', 'r')
print(f.readline())
f.seek(0)
print(f.readline())
f.tell()
```

Write file

```python
f = open('huh.txt', 'w+')
```

```python
lung = '''yay mybro
no
'''
f.write(lung)
```

```python
f.flush()
```

```python
f = open('huh.txt', 'w+')
lines = ['yay mybro', 'no']
# f.writelines(['yay mybro', 'no'])
f.writelines(list(map(lambda l: l+'\n', lines)))
f.flush()
```

Close file

```python
f.close()
```

context manager

```python
# move things in hello to huh
with open('huh.txt', 'a+') as my_file:
    with open('hello.txt', 'r') as bro_file:
        buffer = bro_file.read()
        my_file.write(buffer)
```

```python
with open('huh.txt', 'a+') as my_file:
    my_file.write('Kanplu not plays game in class\n')
    my_file.seek(0)
    print(my_file.read())
```
