```python
import math

class RectangularCoordinates:
    
    def __init__(self, x: int, y: int):
        if not isinstance(x, int) or not isinstance(y, int):
            raise TypeError("Pls enter int plssssss")
        self.x = x
        self.y = y
    def get_coordinates(self) -> tuple[int, int]:
            return (self.x, self.y)

    def is_on_axis(self) -> bool:
        
        return self.x == 0 or self.y == 0

    def get_quadrant(self) -> str:
        
        if self.x > 0 and self.y > 0:
            return "Quadrant 1"
        elif self.x < 0 and self.y > 0:
            return "Quadrant 2"
        elif self.x < 0 and self.y < 0:
            return "Quadrant 3"
        elif self.x > 0 and self.y < 0:
            return "Quadrant 4"
        
    def distance_from_x_axis(self) -> int:
        
        return abs(self.y)

    def distance_from_y_axis(self) -> int:
        
        return abs(self.x)

    def distance_to_origin(self) -> float:
        
        return math.sqrt(self.x**2 + self.y**2)\
        
    def get_midpoint(self, a: int, b: int) -> tuple[float, float]:
        return math.sqrt(self.x + a)/2, (self.y + b)/2

    def __repr__(self):
        return f"RectangularCoordinates(x={self.x}, y={self.y})"
    
# print('yayayay but I can\'t do adv')
```
