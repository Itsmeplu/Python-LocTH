```python
from sklearn.model_selection import train_test_split
```

```python
training_data, test_data = train_test_split(df, train_size=0.8)
X_train, Y_train = training_data.drop('G3', axis=1), training_data['G3']
X_test, Y_test = test_data.drop('G3', axis=1), test_data['G3']
print(X_train, Y_train)
```

## Loss Funtion

$$ MSE= \frac{1}{n} \sum {(\hat{y}-y)^2}$$

```python
def calculate_mse(y_true, y_pred):
    mse = np.mean((y_true - y_pred) ** 2)
    return mse
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
model.coef_, model.intercept_ # coef is weight and intercept is bias
```

## Test

```python
y_hat = model.predict(X_test)
```

```python
mse = calculate_mse(Y_test, y_hat)
print("MSE", mse)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
print(mean_absolute_error(Y_test, y_hat))
print(mean_squared_error(Y_test, y_hat))
print(r2_score(Y_test, y_hat))
```

```python
y_hat = model.predict(X_test)
plt.scatter(Y_test, y_hat)
plt.xlabel('Actual Grade')
plt.ylabel('Predicted Grade')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--') 
```
